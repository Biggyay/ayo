```python
from datetime import datetime

def extract_and_map_month(date):
    
    date_column = pd.to_datetime(date, format='%m/%d/%Y %I:%M:%S %p')

    month_names = date_column.dt.month_name()

    return month_names
```

```python
df['Month'] = extract_and_map_month(df['Date'])
```

```python
# Correlation between numerical variables
# Select numerical columns
num_cols = ['Community Area', 'X Coordinate', 'Y Coordinate', 'Latitude', 'Longitude']

# Compute correlation matrix
corr_matrix = df[num_cols].corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
# Correlation between crime types and location:
location_crime_cross = pd.crosstab(df['Crime Type'], df['Location'])

# Plotting heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(location_crime_cross, cmap='coolwarm', annot=True, fmt='d')
plt.title('Cross-tabulation of Crime Type and Location')
plt.xlabel('Location')
plt.ylabel('Crime Type')
plt.show()
```

```python
# Correlation between crime rate and temporal factors:
# Selecting relevant columns
temporal_cols = ['Crime Rate', 'Time of Day', 'Day of Week', 'Month']

# Compute correlation matrix
temporal_corr = df[temporal_cols].corr()

# Plotting heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(temporal_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation between Crime Rate and Temporal Factors')
plt.show()
```

```python
# Correlation between crime rate and law enforcement activity:
# Box plot for Crime Rate and Law Enforcement Activity
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Crime Rate', 'Arrests']])
plt.title('Box Plot of Crime Rate and Law Enforcement Activity')
plt.xticks(rotation=45)
plt.show()
```
